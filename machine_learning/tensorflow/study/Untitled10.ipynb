```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
```

```python
#このモジュールを用いて、MNISTのデータセットをダウンロードして、オブジェクトに格納する。
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)
```

```python
#変数mnistに格納されたオブジェクトのメソッドを利用して、データを取り出すことが出来る。
#例えば、次は、トレーニングセットから10個分のデータを取り出す。
images, labels = mnist.train.next_batch(10)
```

```python
#取り出した画像データは、28×28=784個のピクセルについて、それぞれの濃度を表す数値を並べたりすとになっている。
#例えば、次のコマンドで取り出した画像データの中から、1つ目にデータの中身を表示することができる。
print images[0]
```

```python
#同様に、対応するラベルのデータを表示すると、次の用になる。
#1-of-Kベクトルでこの画像は7を示す。
print labels[0]
```

```python
#最後に10個分の画像の表示をする。
fig = plt.figure(figsize=(8,4))
for c, (image, label) in enumerate(zip(images, labels)):
    subplot = fig.add_subplot(2,5,c+1)
    subplot.set_xticks([])
    subplot.set_yticks([])
    subplot.set_title('%d' % np.argmax(label))
    subplot.imshow(image.reshape((28,28)), vmin=0, vmax=1,
                   cmap=plt.cm.gray_r, interpolation="nearest")
plt.show()
```
